# Data visualization in Jupyter Notebook
As we have seen to this point, Agate has some built in charts: Bar charts and histograms using bins. But what if we need more? What if we want to make our own? 

Python has several very powerful charting libraries, but one stands above all: Matplotlib. It's a mathematical/statistical/scientific charting library, and the possibilities within it are so daunting that there are actually libraries written over top of Matplotlib to make it easier. One of the most interesting is Seaborn. Let's use Seaborn to make some charts. 

First, like any Python project, we need to import the libraries we'll need. This import will get just about everything you'll need, so keep it around for repeated use.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set(style="ticks")
%matplotlib inline
```

For the purposes of this assignment, I've created a data file called toplots.csv. It's the 20 most ticketed parking lots at UNL over four years. To get that, I used Agate to create a CSV of those top lots. If you want to see how it was done, [it's in the repository](https://github.com/mattwaite/JOUR407-Data-Journalism/blob/master/Assignments/ChartAssignment/TopTicketTable.ipynb).

```python
df = pd.read_csv('../../Data/toplots.csv')
```

Now on to the Seaborn parts. [The documentation for Seaborn](http://web.stanford.edu/~mwaskom/software/seaborn/index.html) contains a ton of examples and a blizzard of options. Something to keep in mind: Seaborn is the easy one. It's an abstraction library written on top of Matplotlib to make it easier to create charts quickly. So as complex as it is, it's a fraction of what Matplotlib can do.

To make a simple bar chart, you need to use Seaborn (shortened to sns) and barplot. Barplot takes some minimal input: the x axis, which is our count; the y axis, which in this case is Location, what data file you're working with and that's about it. We'll give it a color too, with b for blue.

```python
sns.barplot(x="count", y="Location", data=df, color="b")
```

Hmmm. Not a fan of the standard output. Why have all these lines and tickmarks all over. Happily, Seaborn has a simple command to get rid of those, calle despine, and you can specify which ones you want to get rid of.

```python
sns.barplot(x="count", y="Location", data=df, color="b")
sns.despine(left=True, bottom=True)
```

Hmmm. Still not a fan. The blue is all wrong.

```python
sns.set_color_codes("dark")
sns.barplot(x="count", y="Location", data=df, color="b")
sns.despine(left=True, bottom=True)
```

Still don't like all the tickmarks. Well, Seaborn has some default styles that'll get you pretty close to what you want right out of the box. It's called set_style and the one that comes closest to regular journalistic chart output is called whitegrid.

```python
sns.set_style("whitegrid")
sns.set_color_codes("dark")
sns.barplot(x="count", y="Location", data=df, color="b")
```
